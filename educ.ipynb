prevendo como os países possivelmente se comportaram no futuro

```python
def conquista_Pr(DataFrame):
  pais = DataFrame.iloc[1,3]
  X = DataFrame['Time'].values.reshape(-1, 1)
  y = DataFrame['Educational attainment, at least completed upper secondary, population 25+, total (%) (cumulative) [SE.SEC.CUAT.UP.ZS]'].values
  y = np.nan_to_num(y, nan=np.nanmean(y))
  model = LinearRegression()
  model.fit(X, y)
  y_pred = model.predict(X)
  X_future = np.arange(X[-1][0] + 1, X[-1][0] + 6).reshape(-1, 1)
  y_future_pred = model.predict(X_future)
  plt.figure(figsize=(10,3))
  plt.scatter(X, y, color='blue', label='Conquista educacional real')
  plt.plot(X, y_pred, color='red', label='Regression Line')
  plt.plot(X_future, y_future_pred, color='green', label='Previsão futura')
  plt.xlabel('Ano')
  plt.ylabel('Conquista educacional')
  plt.title(f'Modelo de regressão para {pais} (Conquista educacional)')
  plt.grid(True)
  plt.legend()
  plt.show()
```

```python
conquista_Pr(DataFrame = che)
```
